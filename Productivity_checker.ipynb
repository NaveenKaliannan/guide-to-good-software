```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.DataFrame(   
  {"Date":['20-09-2022'],
   "SP_achieved":[79],
   "SP_ideal":[70],
   "SP_created":[69]})

df['CUMSUM_SP_achieved'] = df['SP_achieved'].cumsum()
df['CUMSUM_SP_ideal'] = df['SP_ideal'].cumsum()
df['CUMSUM_SP_created'] = df['SP_created'].cumsum()
```

```python
print(df)
```

```python
plt.figure()
plt.plot(df.index,df['CUMSUM_SP_achieved'],color='red',label='Story points achieved')
plt.plot(df.index,df['CUMSUM_SP_ideal'],color='black',label='Reference')
plt.plot(df.index,df['CUMSUM_SP_created'],color='green',label='Story points created')
plt.xlabel('Time (in 2 weeks)')    
plt.ylabel('Story points (in numbers)') 
plt.title('Productivity') 
plt.legend(loc='best')
plt.show()  
```
